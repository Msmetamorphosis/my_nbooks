```python
ramen_ratings['Stars'] = pd.to_numeric(ramen_ratings['Stars'], errors='coerce')
highest_rated_countries = ramen_ratings.groupby('Country')['Stars'].mean().sort_values(ascending=False).head(3).reset_index()
highest_rated_countries.style.hide(axis="index")
```

```python
lowest_rated_countries = ramen_ratings.groupby('Country')['Stars'].mean().sort_values().head(3).reset_index()
lowest_rated_countries.style.hide(axis="index")
```
